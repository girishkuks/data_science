```python
data.rename(columns = {'Salary (Target)': 'Salary'}, inplace = True)
```

```python
data['Salary'] = data['Salary'].str.replace(',','').astype(float)
```

```python
data.head()
```

```python
data.corr()
```

```python
X = data.drop(['Salary'], axis = 1)
Y = data['Salary']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scalar = StandardScaler()

X_scaled = scalar.fit_transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(X_scaled,Y,random_state=42, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

from sklearn.linear_model import LinearRegression

# Initialize and fit the model
model = LinearRegression()
model.fit(X_train, Y_train)


from sklearn.linear_model import Lasso

# Initialize the Lasso model with a chosen alpha value
lasso = Lasso(alpha=1.0)
lasso.fit(X_train, Y_train)

# Retrieve and print Lasso coefficients
print("Lasso Coefficients:", lasso.coef_)

from sklearn.metrics import mean_squared_error, r2_score

# Linear regression predictions
y_pred_linear = model.predict(X_test)
mse_linear = mean_squared_error(Y_test, y_pred_linear)
r2_linear = r2_score(Y_test, y_pred_linear)

# Lasso predictions
y_pred_lasso = lasso.predict(X_test)
mse_lasso = mean_squared_error(Y_test, y_pred_lasso)
r2_lasso = r2_score(Y_test, y_pred_lasso)

print("Linear Regression MSE:", mse_linear, "R^2:", r2_linear)
print("Lasso Regression MSE:", mse_lasso, "R^2:", r2_lasso)
```

```python
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```
